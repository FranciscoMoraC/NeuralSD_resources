## Test/validation split

```python
def remove_constant_columns(df_train, df_test):
    cols_to_remove = []
    for c in df_train.columns:
        if df_train[c].nunique() == 1 or df_test[c].nunique() == 1:
            print("Column", c, "contains only one unique value. Removing...")
            cols_to_remove.append(c)
    df_train = df_train.drop(cols_to_remove, axis=1)
    df_test = df_test.drop(cols_to_remove, axis=1)
    return df_train, df_test
```

```python
# Mushroom
mushroom = pd.read_csv('../data/agaricus-lepiota-train.csv')
mushroom_tuning, mushroom_validation = train_test_split(mushroom, test_size=0.2, random_state=42)
# mushroom_tuning, mushroom_validation = remove_constant_columns(mushroom_tuning, mushroom_validation)
mushroom_tuning.to_csv('../data/agaricus-lepiota-tuning.csv', index=False)
mushroom_validation.to_csv('../data/agaricus-lepiota-validation.csv', index=False)
# Soybean
soybean = pd.read_csv('../data/soybean_train.csv')
soybean_tuning, soybean_validation = train_test_split(soybean, test_size=0.2, random_state=42)
soybean_tuning, soybean_validation = remove_constant_columns(soybean_tuning, soybean_validation)
soybean_tuning.to_csv('../data/soybean-tuning.csv', index=False)
soybean_validation.to_csv('../data/soybean-validation.csv', index=False)
# Chess
chess = pd.read_csv('../data/chess_train.csv')
chess_tuning, chess_validation = train_test_split(chess, test_size=0.2, random_state=42)
chess_tuning, chess_validation = remove_constant_columns(chess_tuning, chess_validation)
chess_tuning.to_csv('../data/chess-tuning.csv', index=False)
chess_validation.to_csv('../data/chess-validation.csv', index=False)
# Connect-4
connec4 = pd.read_csv('../data/connect-4_train.csv')
connec4_tuning, connec4_validation = train_test_split(connec4, test_size=0.2, random_state=42)
connec4_tuning, connec4_validation = remove_constant_columns(connec4_tuning, connec4_validation)
connec4_tuning.to_csv('../data/connect-4-tuning.csv', index=False)
connec4_validation.to_csv('../data/connect-4-validation.csv', index=False)
```

# Student Success Prediction

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
student = pd.read_csv('../data/StudentDropout.csv', sep = ";")
# Remove "Age of enrollment" column and float64 columns
cols_to_remove = ["Age at enrollment"]
for c in student.columns:
    if student[c].dtype == "float64":
        cols_to_remove.append(c)
student = student.drop(cols_to_remove, axis=1)
# Rename columns with blank spaces
student.columns = student.columns.str.replace(" ", "_")
student = student.astype("str")
student_train, student_test = train_test_split(student, test_size=0.2, random_state=42)
student_tuning, student_validation = train_test_split(student_train, test_size=0.2, random_state=42)
column_order = student.columns.to_list()
column_order = [column_order[-1]] + column_order[:-1]
student = student[column_order]
student_train = student_train[column_order]
student_test = student_test[column_order]
student_tuning = student_tuning[column_order]
student_validation = student_validation[column_order]
student_train.to_csv('../data/student_train.csv', index=False)
student_test.to_csv('../data/student_test.csv', index=False)
student_tuning.to_csv('../data/student-tuning.csv', index=False)
student_validation.to_csv('../data/student-validation.csv', index=False)
student.to_csv('../data/student.csv', index=False)
```
