```python
import os
import pandas as pd
from itertools import product
from tqdm import tqdm
```

# Facebook

```python
dataset = "Facebook"

df = pd.read_csv(f"../data/facebook.csv")

best_params = None
max_youden = -1
```

```python
target_tuple = ("target", "government")
target_entry = df[target_tuple[0]] == target_tuple[1]
```

```python
interpretability_df = pd.DataFrame(columns=["Dataset", "# Subgroups", "Avg size", "Maximum Jaccard"])
```

```python
dir = f"results/{dataset}"
file_path = dir + "/" + os.listdir(dir)[0] + "/result.log"
results_df = pd.read_csv(file_path, sep="\t")
descriptions = results_df["Description"].tolist()
descriptions = [[pat.replace("'", "\"") for pat in eval(desc)] for desc in descriptions]
global_entry = pd.Series(False, index=df.index)
n_subgroups = len(descriptions)
sizes = []
prev_entries = []
max_jaccard = 0
for desc in descriptions:
    entry = pd.Series(True, index=df.index)
    for pat in desc:
        entry &= (df[pat] == 1)
    sizes.append(len(desc))
    for e in prev_entries:
        jaccard = sum(entry & e)/sum(entry | e)
        max_jaccard = max(max_jaccard, jaccard)
    prev_entries.append(entry)
    global_entry |= entry
positive = sum(global_entry & target_entry)/sum(target_entry)
negative = sum(global_entry & (~target_entry))/sum(~target_entry)
youden = positive - negative
interpretability_df.loc[len(interpretability_df)] = [dataset, n_subgroups, sum(sizes)/len(sizes), max_jaccard]
```

```python
print(youden)
```
