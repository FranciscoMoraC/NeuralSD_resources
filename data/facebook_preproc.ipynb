```python
import pandas as pd
```

```python
df = pd.read_csv("./musae_facebook_edges.csv", sep =",")
df["id_2"].nunique()
```

```python
import json
# f = open("./musae_facebook_features.json", "r")
json_data = json.load(open("./musae_facebook_features.json", "r"))
liked = set()
for v in json_data.values():
    for k in v:
        liked.add(k)
```

```python
len(liked)
```

```python
from tqdm import tqdm
```

```python
df = pd.DataFrame(columns=list(liked), index = json_data.keys())
for k, v in tqdm(json_data.items()):
    for l in liked:
        if l in v:
            df.at[k, l] = 1
        else:
            df.at[k, l] = 0
```

```python
df_target = pd.read_csv("./musae_facebook_target.csv", sep =",")
df_target.head()
```

```python
df.index = df.index.astype("int")
df.sort_index(inplace=True)
```

```python
(df.index == df_target["id"].values).all()
```

```python
df["target"] = df_target["page_type"]
```

```python
df.to_csv("./facebook.csv", index=False)
```
