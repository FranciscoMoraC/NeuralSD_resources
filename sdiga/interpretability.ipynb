```python
def get_interpretability_measures(df, results_path):
    with open(results_path, "r") as f:
        lines = f.readlines()

    descriptions = []
    for line in lines:
        if line.startswith("Description:"):
            desc = line.split("Description: [")[1].split("], Target:")[0].strip()
            descriptions.append(desc)

    n_subgroups = 0
    sizes = []
    prev_entries = []
    max_jaccard = 0
    for desc in descriptions:
        n_subgroups += 1
        entry = pd.Series([True]*len(df))
        size = 0
        for sel in desc.split(", "):
            attribute, value = sel.split(" = ")
            attribute = attribute.strip()
            value = value.strip().strip("'")
            entry &= (df[attribute] == value)
            size += 1
        sizes.append(size)
        for prev_entry in prev_entries:
            jaccard = sum(entry & prev_entry)/sum(entry | prev_entry)
            max_jaccard = max(max_jaccard, jaccard)
        prev_entries.append(entry)
    return n_subgroups, sum(sizes)/n_subgroups, max_jaccard
```

```python
facebook = get_facebook_data()
facebook_results = "results_SDIGA_21570_Facebook.txt"
n_subgroups, avg_length, max_jaccard = get_interpretability_measures(facebook, facebook_results)
print("Facebook:")
print("Number of subgroups:", n_subgroups)
print("Average subgroup description length:", avg_length)
print("Max Jaccard similarity between subgroups:", max_jaccard)
```
