```python
import pandas as pd
```

```python
df = pd.DataFrame(
    columns=[
        "hidden_dim",
        "epochs",
        "weight_decay",
        "lr",
        "batch_size",
        "youden",
    ]
)

selected_df = pd.DataFrame(
    columns=["Dataset", "hidden_dim", "epochs", "weight_decay", "lr", "batch_size", "youden"]
)
```

```python
for line in open("../code/facebook_results.txt"):
    if not line.startswith("("):
        continue
    youden = float(line.split(" ")[-1])
    params = "".join(line.split(" ")[:-1])
    hidden_dim, epochs, weight_decay, lr, batch_size = eval(params)
    df.loc[len(df)] = [hidden_dim, epochs, weight_decay, lr, batch_size, youden]
df.sort_values("youden", ascending=False, inplace=True)
df.head(10)
selected = df.iloc[0]
selected["Dataset"] = "Facebook"
selected_df.loc[len(selected_df)] = selected
```

```python
df = pd.DataFrame(
    columns=[
        "hidden_dim",
        "epochs",
        "weight_decay",
        "lr",
        "batch_size",
        "youden",
    ]
)

prev_selected = (2000,25,5.0,0.005, 32)
youden_prev_selected = None

for line in open("../code/synthetic_results.txt"):
    if not line.startswith("("):
        continue
    youden = float(line.split(" ")[-1])
    params = "".join(line.split(" ")[:-1])
    hidden_dim, epochs, weight_decay, lr, batch_size = eval(params)
    if (hidden_dim, epochs, weight_decay, lr, batch_size) == prev_selected:
        print("Previously selected, youden:", youden)
        youden_prev_selected = youden
    df.loc[len(df)] = [hidden_dim, epochs, weight_decay, lr, batch_size, youden]
df.sort_values("youden", ascending=False, inplace=True)
df.head(10)
selected = df.iloc[0]
selected["Dataset"] = "Synthetic"
selected_df.loc[len(selected_df)] = selected
if youden_prev_selected is not None and youden_prev_selected == selected["youden"]:
    print("The previously selected hyperparameters are still the best.")
```
