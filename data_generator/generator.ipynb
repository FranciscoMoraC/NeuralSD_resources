```python
import pandas as pd
import numpy as np
import random
```

```python
# Generate list of length n with values 1 and 0 
def generate_list(n, alpha=0.99):
    if alpha == 1:
        return [1] * n
    # Generate a list of 1s and 0s
    lst = [1] * (int(n * alpha)+1) + [0] * (int(n * (1-alpha))+1)
    
    # Shuffle the list to randomize the order
    random.shuffle(lst)
    
    # Adjust the length of the list to n
    lst = lst[:n]
    return lst
```

```python
def expand_csv(input_file, additional_cols=0, additional_cols_values=2, alpha=0.99):
    df = pd.read_csv(input_file, sep=";")
    df = df.fillna(0)

    cols = list(df.columns)
    cols.remove("Index")
    for col in cols:
        df[col] = df[col].astype(int)

    data = { c : [] for c in cols }

    for _, row in df.iterrows():
        try:
            start, end = map(int, row["Index"].split("-"))
        except ValueError:
            continue

        count = end - start
        for col in cols:
            if row[col] == 1:
                lst = generate_list(count, alpha)
                data[col].extend(lst)
            else:
                lst = [0] * count
                data[col].extend(lst)

    df_expanded = pd.DataFrame(data)

    if additional_cols == 0:
        return df_expanded
    new_cols = []
    new_col_names = []
    for i in range(additional_cols):
        col_name = f"col_{i+1}"
        lst = random.choices(range(additional_cols_values), k=len(df_expanded))
        new_cols.append(pd.Series(lst, name=col_name))
        new_col_names.append(col_name)
    df_expanded = pd.concat([df_expanded] + new_cols, axis=1, ignore_index=True)
    # Renombrar las columnas
    df_expanded.columns = cols + new_col_names
    return df_expanded
```

```python
input_file = "input3.csv"
df_expanded = expand_csv(input_file, additional_cols=2000, additional_cols_values=5, alpha=1)
```
