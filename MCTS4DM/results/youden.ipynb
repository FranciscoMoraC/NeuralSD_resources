```python
import pandas as pd
```

```python
results = pd.read_csv("./Facebook/resXP1757932302198/result.log", sep="\t")
```

```python
data = pd.read_csv("../../data/facebook.csv")
```

```python
target_tuple = ("target", "government")
target_entry = (data[target_tuple[0]] == target_tuple[1])
```

```python
global_entry = pd.Series(False, index=data.index)
for i, row in results.iterrows():
    entry = pd.Series(True, index=data.index)
    pat = eval(row["Description"].strip())
    for c in pat:
        entry &= (data[c] == 1)
    global_entry |= entry
```

```python
positive = (target_entry & global_entry).sum()/target_entry.sum()
negative = ((~target_entry) & global_entry).sum()/(~target_entry).sum()
print("Youden", positive - negative)
```
