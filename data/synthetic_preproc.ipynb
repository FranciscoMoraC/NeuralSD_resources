```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("synthetic3.csv", sep = ";")
```

```python
# Randomize the rows of the dataframe
df = df.sample(frac=1, random_state = 42).reset_index(drop=True)
df
```

```python
df_train, df_test = train_test_split(df, test_size=0.2, random_state=42, stratify=df['Target'])
df_train.to_csv("synthetic3_train.csv", index = False)
df_test.to_csv("synthetic3_test.csv", index = False)
df.to_csv("synthetic3_all.csv", index = False)
```
