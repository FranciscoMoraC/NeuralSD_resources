```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("brca_n.csv")
target_tuple = ("Target", "1")
```

```python
df_train, df_test = train_test_split(df, test_size=0.2, random_state=42, stratify=df[ target_tuple[0] ])
df_train.to_csv("brca_n_train.csv", index=False)
df_test.to_csv("brca_n_test.csv", index=False)
```
