```python
import os
import pandas as pd
from itertools import product
from tqdm import tqdm
```

# Facebook

```python
dataset = "Facebook"

grid_parameters = {
    "redundancyStrategy": ["jaccardSupportDescription", "jaccardSupportDescriptionTarget", "sumJaccard"],
    "maxRedundancy": [0.4, 0.6, 0.8],
    "nbIter" : [1000, 10000, 100000],
    "measure" : ["WRAcc", "WeightedRelativeF1", "RelativeF1", "Acc"]
}

df_train = pd.read_csv(f"../data/facebook_train.csv")
df_test = pd.read_csv(f"../data/facebook_test.csv")

best_params = None
max_youden = -1
```

```python
target_tuple = ("target", "government")
target_entry = df_test[target_tuple[0]] == target_tuple[1]
```

```python
for redundancyStrategy, maxRedundancy, nbIter, measure in tqdm(product(*grid_parameters.values())):
    # f = open(f"results/{dataset}_{redundancyStrategy}_{maxRedundancy}_{nbIter}_{measure}/result.log", "r")
    dir = f"results/{dataset}_{redundancyStrategy}_{maxRedundancy}_{nbIter}_{measure}"
    # print(os.listdir(dir))
    file_path = dir + "/" + os.listdir(dir)[0] + "/result.log"
    results_df = pd.read_csv(file_path, sep="\t")
    # print(f"Params: {redundancyStrategy}, {maxRedundancy}, {nbIter}, {measure}")
    # display(results_df)
    descriptions = results_df["Description"].tolist()
    descriptions = [[pat.replace("'", "\"") for pat in eval(desc)] for desc in descriptions]
    global_entry = pd.Series(False, index=df_test.index)
    for desc in descriptions:
        entry = pd.Series(True, index=df_test.index)
        for pat in desc:
            entry &= (df_test[pat] == 1)
        global_entry |= entry
    positive = sum(global_entry & target_entry)/sum(target_entry)
    negative = sum(global_entry & (~target_entry))/sum(~target_entry)
    youden = positive - negative
    if youden > max_youden:
        max_youden = youden
        best_params = (redundancyStrategy, maxRedundancy, nbIter, measure)
```

```python
print(max_youden)
best_params
```
